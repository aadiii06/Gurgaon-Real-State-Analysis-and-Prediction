```python
import numpy as np
import pandas as pd
```

```python
flats = pd.read_csv('flats_cleaned.csv')
houses = pd.read_csv('house_cleaned.csv')
```

```python
# Combines the 'flats' and 'houses' DataFrames into a single DataFrame and resets the index
df = pd.concat([flats,houses],ignore_index=True)
```

```python
# Shuffles the rows of the DataFrame randomly and resets the index to maintain sequential order
df = df.sample(df.shape[0],ignore_index=True)
```
